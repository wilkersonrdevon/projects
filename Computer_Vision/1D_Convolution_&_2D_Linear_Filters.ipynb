```python
#Required to use the imshow command within colab
from google.colab.patches import cv2_imshow

#Numpy, Matplotlib and OpenCV are required
import cv2
import matplotlib.pyplot as plt
import numpy as np


from scipy import misc
#load the staircase image
ascent = misc.ascent() 

cv2_imshow(ascent)
print(ascent.shape)

#retrieve the first row
first_row = ascent[1]
#create the 1x2 finite difference filter
filter = np.array([1,-1])
#filter the row
filtered_row = np.convolve(first_row, filter)
#plot the image intensity and the filter output
plt.plot(first_row)
plt.xlabel("Index")
plt.ylabel("Intensity")
plt.plot(filtered_row)

```

```python
from scipy import signal
from scipy import misc
import matplotlib.pyplot as plt

ascent=misc.ascent()

#create the numpy filters
horizontalEdgeFilter = np.array([[-1,0,1],
                                 [-1,0,1],
                                 [-1,0,1]])
verticalEdgeFilter = np.array([[1,1,1],
                               [0,0,0],
                               [-1,-1,-1]])
#apply them using signal.convolve2d
horizontalGradient = signal.convolve2d(ascent, horizontalEdgeFilter)
verticalGradient = signal.convolve2d(ascent, verticalEdgeFilter)
#display the original imaage and the filtered images
cv2_imshow(ascent)
cv2_imshow(horizontalGradient)
cv2_imshow(verticalGradient)


```
