```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
parrot = cv2.imread("single-parrot-small.png")
parrot_group = cv2.imread("group-parrot-small.png")
vacuum = cv2.imread("eureka-vacuum-small.png")

cv2_imshow(parrot)
cv2_imshow(parrot_group)
cv2_imshow(vacuum)

parrot_BGR = cv2.split(parrot)
group_parrot_BGR = cv2.split(parrot_group)
vacuum_BGR = cv2.split(vacuum)
hist_size = 256
hist_range = [0, 256]
accumulate = False

b_parrot_hist = cv2.calcHist(parrot_BGR, [0], None, [
                             hist_size], hist_range, accumulate=accumulate)
g_parrot_hist = cv2.calcHist(parrot_BGR, [1], None, [
                             hist_size], hist_range, accumulate=accumulate)
r_parrot_hist = cv2.calcHist(parrot_BGR, [2], None, [
                             hist_size], hist_range, accumulate=accumulate)
b_group_parrot_hist = cv2.calcHist(group_parrot_BGR, [0], None, [
                                   hist_size], hist_range, accumulate=accumulate)
g_group_parrot_hist = cv2.calcHist(group_parrot_BGR, [1], None, [
                                   hist_size], hist_range, accumulate=accumulate)
r_group_parrot_hist = cv2.calcHist(group_parrot_BGR, [2], None, [
                                   hist_size], hist_range, accumulate=accumulate)
b_vacuum_hist = cv2.calcHist(vacuum_BGR, [0], None, [
                             hist_size], hist_range, accumulate=accumulate)
g_vacuum_hist = cv2.calcHist(vacuum_BGR, [1], None, [
                             hist_size], hist_range, accumulate=accumulate)
r_vacuum_hist = cv2.calcHist(vacuum_BGR, [2], None, [
                             hist_size], hist_range, accumulate=accumulate)
hist_w = 512
hist_h = 400
bin_w = int(round(hist_w/hist_size))

parrot_hist = np.zeros((hist_h, hist_w, 3), dtype=np.uint8)

cv2.normalize(b_parrot_hist, b_parrot_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(g_parrot_hist, g_parrot_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(r_parrot_hist, r_parrot_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)

for i in range(1, hist_size):
    cv2.line(parrot_hist, (bin_w*(i-1), hist_h - round((int)(b_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(b_parrot_hist[i]))),
             (255, 0, 0), thickness=2)
    cv2.line(parrot_hist, (bin_w*(i-1), hist_h - round((int)(g_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(g_parrot_hist[i]))),
             (0, 255, 0), thickness=2)
    cv2.line(parrot_hist, (bin_w*(i-1), hist_h - round((int)(r_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(r_parrot_hist[i]))),
             (0, 0, 255), thickness=2)

group_parrot_hist = np.zeros((hist_h, hist_w, 3), dtype=np.uint8)

cv2.normalize(b_group_parrot_hist, b_group_parrot_hist,
              alpha=0, beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(g_group_parrot_hist, g_group_parrot_hist,
              alpha=0, beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(r_group_parrot_hist, r_group_parrot_hist,
              alpha=0, beta=hist_h, norm_type=cv2.NORM_MINMAX)

for i in range(1, hist_size):
    cv2.line(group_parrot_hist, (bin_w*(i-1), hist_h - round((int)(b_group_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(b_group_parrot_hist[i]))),
             (255, 0, 0), thickness=2)
    cv2.line(group_parrot_hist, (bin_w*(i-1), hist_h - round((int)(g_group_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(g_group_parrot_hist[i]))),
             (0, 255, 0), thickness=2)
    cv2.line(group_parrot_hist, (bin_w*(i-1), hist_h - round((int)(r_group_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(r_group_parrot_hist[i]))),
             (0, 0, 255), thickness=2)

vacuum_hist = np.zeros((hist_h, hist_w, 3), dtype=np.uint8)

cv2.normalize(b_vacuum_hist, b_vacuum_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(g_vacuum_hist, g_vacuum_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(r_vacuum_hist, r_vacuum_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)

for i in range(1, hist_size):
    cv2.line(vacuum_hist, (bin_w*(i-1), hist_h - round((int)(b_vacuum_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(b_vacuum_hist[i]))),
             (255, 0, 0), thickness=2)
    cv2.line(vacuum_hist, (bin_w*(i-1), hist_h - round((int)(g_vacuum_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(g_vacuum_hist[i]))),
             (0, 255, 0), thickness=2)
    cv2.line(vacuum_hist, (bin_w*(i-1), hist_h - round((int)(r_vacuum_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(r_vacuum_hist[i]))),
             (0, 0, 255), thickness=2)

cv2_imshow(parrot_hist)
cv2_imshow(group_parrot_hist)
cv2_imshow(vacuum_hist)

hist_size = 8

b_parrot_hist = cv2.calcHist(parrot_BGR, [0], None, [
                             hist_size], hist_range, accumulate=accumulate)
g_parrot_hist = cv2.calcHist(parrot_BGR, [1], None, [
                             hist_size], hist_range, accumulate=accumulate)
r_parrot_hist = cv2.calcHist(parrot_BGR, [2], None, [
                             hist_size], hist_range, accumulate=accumulate)
b_group_parrot_hist = cv2.calcHist(group_parrot_BGR, [0], None, [
                                   hist_size], hist_range, accumulate=accumulate)
g_group_parrot_hist = cv2.calcHist(group_parrot_BGR, [1], None, [
                                   hist_size], hist_range, accumulate=accumulate)
r_group_parrot_hist = cv2.calcHist(group_parrot_BGR, [2], None, [
                                   hist_size], hist_range, accumulate=accumulate)
b_vacuum_hist = cv2.calcHist(vacuum_BGR, [0], None, [
                             hist_size], hist_range, accumulate=accumulate)
g_vacuum_hist = cv2.calcHist(vacuum_BGR, [1], None, [
                             hist_size], hist_range, accumulate=accumulate)
r_vacuum_hist = cv2.calcHist(vacuum_BGR, [2], None, [
                             hist_size], hist_range, accumulate=accumulate)

parrot_hist_2 = np.zeros((hist_h, hist_w, 3), dtype=np.float32)

cv2.normalize(b_parrot_hist, b_parrot_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(g_parrot_hist, g_parrot_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(r_parrot_hist, r_parrot_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)

for i in range(1, hist_size):
    cv2.line(parrot_hist_2, (bin_w*(i-1), hist_h - round((int)(b_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(b_parrot_hist[i]))),
             (8, 0, 0), thickness=2)
    cv2.line(parrot_hist_2, (bin_w*(i-1), hist_h - round((int)(g_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(g_parrot_hist[i]))),
             (0, 8, 0), thickness=2)
    cv2.line(parrot_hist_2, (bin_w*(i-1), hist_h - round((int)(r_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(r_parrot_hist[i]))),
             (0, 0, 8), thickness=2)

group_parrot_hist_2 = np.zeros((hist_h, hist_w, 3), dtype=np.float32)

cv2.normalize(b_group_parrot_hist, b_group_parrot_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(g_group_parrot_hist, g_group_parrot_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(r_group_parrot_hist, r_group_parrot_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)

for i in range(1, hist_size):
    cv2.line(group_parrot_hist_2, (bin_w*(i-1), hist_h - round((int)(b_group_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(b_group_parrot_hist[i]))),
             (8, 0, 0), thickness=2)
    cv2.line(group_parrot_hist_2, (bin_w*(i-1), hist_h - round((int)(g_group_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(g_group_parrot_hist[i]))),
             (0, 8, 0), thickness=2)
    cv2.line(group_parrot_hist_2, (bin_w*(i-1), hist_h - round((int)(r_group_parrot_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(r_group_parrot_hist[i]))),
             (0, 0, 8), thickness=2)

vacuum_hist_2 = np.zeros((hist_h, hist_w, 3), dtype=np.float32)

cv2.normalize(b_vacuum_hist, b_vacuum_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(g_vacuum_hist, g_vacuum_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)
cv2.normalize(r_vacuum_hist, r_vacuum_hist, alpha=0,
              beta=hist_h, norm_type=cv2.NORM_MINMAX)

for i in range(1, hist_size):
    cv2.line(vacuum_hist_2, (bin_w*(i-1), hist_h - round((int)(b_vacuum_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(b_vacuum_hist[i]))),
             (8, 0, 0), thickness=2)
    cv2.line(vacuum_hist_2, (bin_w*(i-1), hist_h - round((int)(g_vacuum_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(g_vacuum_hist[i]))),
             (0, 8, 0), thickness=2)
    cv2.line(vacuum_hist_2, (bin_w*(i-1), hist_h - round((int)(r_vacuum_hist[i-1]))),
             (bin_w*(i), hist_h - round((int)(r_vacuum_hist[i]))),
             (0, 0, 8), thickness=2)

parrotCmpParrot = cv2.compareHist(parrot_hist_2, parrot_hist_2, cv2.HISTCMP_INTERSECT)
parrotCmpGroup = cv2.compareHist(parrot_hist_2, group_parrot_hist_2, cv2.HISTCMP_INTERSECT)
parrotCmpVacuum = cv2.compareHist(parrot_hist_2, vacuum_hist_2, cv2.HISTCMP_INTERSECT)
print(parrotCmpParrot, parrotCmpGroup, parrotCmpVacuum)
```
