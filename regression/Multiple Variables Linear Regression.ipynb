```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data_two_variables = pd.read_csv("data_two_variables.csv")
X1 = data_two_variables.iloc[:,0]
X2 = data_two_variables.iloc[:,1]
Y = data_two_variables.iloc[:,2]

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X1, X2, Y)

plt.show()
```

```python
m1 = 0
m2 = 0
m3 = 0

L = 0.001 
epochs = 10000

n = float(len(X1)) 


for i in range(epochs): 
    Y_pred = m1*X1 + m2*X2 + m3
    D_m1 = (-2/n) * sum(X1 * (Y - Y_pred))
    D_m2 = (-2/n) * sum(X2 * (Y - Y_pred))
    D_m3 = (-2/n) * sum(Y - Y_pred) 
    m1 = m1 - L * D_m1
    m2 = m2 - L * D_m2
    m3 = m3 - L * D_m3
print (m1, m2, m3)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X1, X2, Y)
Y_pred = m1*X1 + m2*X2 + m3
ax.scatter(X1, X2, Y_pred, color = 'red')

plt.show()
```
