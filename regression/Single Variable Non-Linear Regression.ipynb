```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nonlinear_data = pd.read_csv("data_nonlinear.csv")
X = nonlinear_data.iloc[:,0]
Y = nonlinear_data.iloc[:,1]

plt.scatter(X,Y)
plt.show()
```

```python
a = 0
b = 0
c = 0

L = 0.0001 
epochs = 10000 

n = float(len(X)) 

for i in range(epochs): 
    Y_pred = a*X**2 + b*X + c
    D_a = (-2/n) * sum(X**2 * (Y - Y_pred))
    D_b = (-2/n) * sum(X * (Y - Y_pred))
    D_c = (-2/n) * sum(Y - Y_pred) 
    a = a - L * D_a
    b = b - L * D_b 
    c = c - L * D_c
print (a, b, c)
```

```python
plt.scatter(X, Y)
plt.scatter(X, a * X**2 + b * X + c, color='red')
plt.show()
```
